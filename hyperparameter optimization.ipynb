```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
```

```python
#loading the dataset
df=pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv')
df.head()
```

```python
#creating the independent and the dependent feature
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import KFold
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor()
```

```python
kfold=KFold(n_splits=5,shuffle=True,random_state=1)
scores=cross_val_score(knn,x,y,cv=kfold,scoring='r2')
```

```python
scores.mean()
```

GridsearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
knn=KNeighborsRegressor()
```

```python
param_grid={
    'n_neighbors':[1,3,5,7,10,12,15,17,20],
    'weights':['uniform','distance'],
    'algorithm':['ball_tree','kd_tree','brute'],
    'p':[1,2]
}
```

```python
gcv=GridSearchCV(knn,param_grid=param_grid,scoring='r2',refit=True,cv=kfold,verbose=2)
```

```python
gcv.fit(x,y)
```

```python
gcv.best_params_
```

```python
gcv.best_score_
```

```python
gcv.cv_results_
```

```python
pd.DataFrame(gcv.cv_results_)[['param_algorithm','param_n_neighbors','param_p','param_weights', 'mean_test_score']].sort_values('mean_test_score',ascending=False)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rcv=RandomizedSearchCV(knn,param_grid,scoring='r2',refit=True,cv=kfold,verbose=2)
```
